```python
df = pd.read_csv('references.csv')
df['topic'] = df['topic'].str.lower()
df['reference'] = df['reference'].str.lower()
df[df['topic'] == 'muffle']
```

```python
def interleave_lists(list1, list2):
    return [x for pair in zip(list1, list2) for x in pair]

def get_unique_preserve_order(lst):
    unique_values = []
    seen = set()

    for item in lst:
        if item not in seen:
            unique_values.append(item)
            seen.add(item)

    return unique_values
```

```python
# all_topics = df['topic'].tolist() + df['reference'].tolist()
all_topics = interleave_lists(df['topic'].tolist(), df['reference'].tolist()) 
topics = get_unique_preserve_order(all_topics)
topic_freq = pd.Series(all_topics).value_counts()
```

```python
topic_freq
```

```python
id = 0
topics_dict = {}
for idx, state in enumerate(topics):
    topics_dict[state] = idx+1
topics_dict
```

```python
{i:j for i, j in enumerate(['topic', 'id', 'group'])}
```

```python
#pd.Series(topics_dict).to_csv('topics.csv')
references_topics = pd.read_csv('reference-topics.csv', names=['topic', 'id', 'group']).drop(0)
references_topics = references_topics.set_index('id')['group'].to_dict()#.rename({i:j for i, j in enumerate(['topic', 'id', 'group'])}, axis=1)#.to_dict()
references_topics
```

```python
random.seed(350)
```

```python
df['Source ID'] =  [topics_dict[state] for state in df['topic']]
df['Target ID'] =  [topics_dict[state] for state in df['reference']]
df1 = pd.DataFrame({'Quantity' : df.groupby( ['Source ID','Target ID'] ).sum()['times']}).reset_index()
df1.head()
```

```python
connections = {}
for index, row in df1.iterrows():
    # Access the values of each column for the current row
    source_id = row['Source ID']
    target_id = row['Target ID']
    
    # Update connections for the source ID
    if source_id in connections:
        connections[source_id].append(target_id)
    else:
        connections[source_id] = [target_id]

    # Update connections for the target ID
    if target_id in connections:
        connections[target_id].append(source_id)
    else:
        connections[target_id] = [source_id]
```

```python
df1['Quantity'] = df1['Quantity'].astype(int)
```

```python
nodes =[]
jsons = {}
links = []
for state,idx in topics_dict.items():
    tmp = {}
    tmp['id']= int(idx)
    tmp['name'] = state
    tmp['grp'] = references_topics[idx]#int(idx % 10)
    tmp['freq'] = int(topic_freq[state])
    tmp['connections'] = [int(x) for x in connections[idx]]
    
    nodes.append(tmp)


for idx, row in df1.iterrows():
    tmp = {}
    # added to reduce number of links
    if row['Source ID'] == row['Target ID']:
        continue
    tmp['source'] = int(row['Source ID'])
    tmp['target'] = int(row['Target ID'])
    tmp['value'] = int(row['Quantity'])
    links.append(tmp)

jsons['nodes'] = nodes
jsons['links'] = links
```

```python
json_data = json.dumps(jsons)
json_data
```

```python
with open('references.json', 'w') as f:
    json.dump(jsons, f)
```
